```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2, 3, 3, 6, 4, 7])
labels = ['JAPAN', 'VIETNAM', 'NEPAL', 'MYANMAR', 'INDONESIA', 'CHINA']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'blue', 'purple']
plt.pie(x, autopct='%d%%', labels=labels, colors=colors)
plt.axis('equal')
plt.title('3M')
plt.show()
```
